```python
def count_zeros(x):
    return len(x[x==0])

def nonzero_mean(x):
    return x[x != 0].mean()

def up_tick(x):
    u = x - x.shift(1)
    return len( u[u>0] )

def down_tick(x):
    u = x - x.shift(1)
    return len( u[u<0] ) 
    
def tick_count(x):
    return len(x)
```

```python
def sma_strategy(df, short, long, thresh = None, neg = 1):
    df[f'sma{short}'] = df.mid.rolling(short).mean()
    df[f'sma{long}'] = df.mid.rolling(long).mean()
    df['sma_signal'] = df[f'sma{short}'] - df[f'sma{long}']
    if thresh == None:
        df['sma_signal'][df['sma_signal']<0] = -1
        df['sma_signal'][df['sma_signal']>0] =  1
    else:
        df['sma_signal'][df['sma_signal']<-thresh] = -1
        df['sma_signal'][df['sma_signal']>thresh] =  1
        df['sma_signal'][ np.abs(df['sma_signal']) != 1] = 0
    df['sma_signal'] = df['sma_signal'] * neg
    return df 
```

```python
def sma_price_strategy(df, sma, thresh = None, neg = 1):
    df[f'sma{sma}'] = df.mid.rolling(sma).mean()
    df['sma_signal'] = df[f'sma{sma}'] - df.mid
    if thresh == None:
        df['sma_signal'][df['sma_signal']<0] = -1
        df['sma_signal'][df['sma_signal']>0] =  1
    else:
        df['sma_signal'][df['sma_signal']<-thresh] = -1
        df['sma_signal'][df['sma_signal']>thresh] =  1
        df['sma_signal'][ np.abs(df['sma_signal']) != 1] = 0
    df['sma_signal'] = df['sma_signal'] * neg
    return df 
```

### Strategy

```python
df = data.copy()
df['trading_hours'] = 1
momentum = 60
holding_period = 480
round_trip = 1.2
spread_cutoff = 0.00002
```

```python
df[f'returns_{momentum}'] = df.mid / df.mid.shift(momentum) - 1
df[f'mom{momentum}'] = df[ f'returns_{momentum}' ]
df[f'mom{momentum}'][ df[f'mom{momentum}'] <= 0 ] = -1
df[f'mom{momentum}'][ df[f'mom{momentum}'] >= 0 ] = 1
#df[f'mom{momentum}'][ np.abs(df[f'mom{momentum}']) != 1 ] = 0
#df[f'mom{momentum}'] *= -1

df.dropna(inplace = True)
df.reset_index(inplace = True)
```

```python
plt.plot( df[f'mom{momentum}'] )
```

```python
df['spread_below_000001'] = 0
df['spread_below_000001'][df.non_zero_spreads <= spread_cutoff ] = 1
```

```python
#df[ df['spread_below_000001'] == 1 ]
```

```python
# clean signal
spread_signal = 'spread_below_000001'
momentum_signal = f'mom{momentum}'
count = 0
for i, j in enumerate( df[spread_signal] ): 
    if np.abs( j ) == 1 and count == 0:
        try:
            df.loc[i, spread_signal] = 0
            df.loc[i+1:i+holding_period, spread_signal] = 1 * df[momentum_signal].loc[i] #* df['trading_hours'].iloc[i] 
            count = holding_period
        except:
            df[spread_signal].loc[i+1:, spread_signal] = 1 * df[momentum_signal].loc[i] #* df['trading_hours'].iloc[i] 
            break
    else:
        if i != 0:
            #print(np.abs( df[spread_signal].loc[i-1] ))
            if np.abs( df[spread_signal].loc[i] ) == 1:
                count -= 1

#df[spread_signal][df.hour >= 20] = 0
#df[spread_signal][df.hour <= 6] = 0
```

```python
pip_profit = df.spread_below_000001 * df.pips * 10000
transaction_costs = np.abs(df.spread_below_000001) / holding_period * round_trip 
```

```python
m = df[(df.spread_below_000001 != 0)].index
mean_pips = pip_profit.loc[m].mean() * holding_period
std_pips = (pip_profit.loc[m] ).std() * np.sqrt( holding_period )
```

```python
#df[ df[spread_signal] != 0 ]
```

```python
print('Pre-transaction costs')
print('Mean pips:', mean_pips)
print('Std pips:', std_pips)
print('Total pips:', pip_profit.sum())
print('\n')
print('After-transaction costs')
print('Mean pips:', (pip_profit - transaction_costs).mean() * holding_period )
print('Total pips:', (pip_profit - transaction_costs).sum())
print('\n')
print('Number of trades:', len(m) / holding_period)
```

```python
plt.plot( df.pips.cumsum() * 10000, label = 'actual pips', alpha = 0.2)
plt.plot( pip_profit.cumsum(), label = 'pre transaction costs')
plt.plot( ( pip_profit - transaction_costs ).cumsum(), label = 'after transaction costs')
plt.legend()
```
