```python
def count_zeros(x):
    return len(x[x==0])

def nonzero_mean(x):
    return x[x != 0].mean()
```

```python
df.columns = ['currency_pair', 'timestamp', 'bid', 'ask' ]
df['all_spreads'] = df.ask - df.bid
df['non_zero_spreads'] = df.ask - df.bid
df['zero_spreads'] = df.ask - df.bid
df['mid'] = ( df.ask + df.bid) / 2
df.index = pd.to_datetime( df.timestamp )
```

```python
df.dtypes
```

```python
aggregation = { 'currency_pair': 'last', 
                'timestamp':'last', 
                'bid': 'last', 
                'ask': 'last',
                'mid': 'last', 
                'all_spreads': 'mean', 
                'non_zero_spreads': nonzero_mean, 
                'zero_spreads': count_zeros }
df = df.resample('1T').agg( aggregation )
df['returns'] = df.mid.pct_change()
df = df[1:]
df['hour'] = [ t.hour for t in df.index ]
df['min'] = [ t.minute for t in df.index ]
df['date'] = [ t.date() for t in df.index ]
```
